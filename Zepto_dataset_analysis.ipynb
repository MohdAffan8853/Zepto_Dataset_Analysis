```python
import pandas as pd

df = pd.read_csv(r"C:\Users\affan\OneDrive\Desktop\Zepto Data Analysis Project\Zepto_data.csv")
df.head(9)
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
# Check missing value

df.isnull().sum()
```

```python
# Remove Duplicate values 

df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
# Data type corrections

df['outOfStock'] = df['outOfStock'].astype(bool)
df['discountPercent'] = df['discountPercent'].astype(int)
df['availableQuantity'] = df['availableQuantity'].astype(int)
df.dtypes
```

```python
# Column name into snake case

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
# Calculate Discount Amount

df['discountamount'] = df['mrp'] - df['discountedsellingprice']
df[['discountamount','discountedsellingprice']].head(10)
```

```python
# Price Per Gram

df['pricepergram'] = df['discountedsellingprice'] / df['weightingms']
df[['pricepergram','discountedsellingprice','weightingms']].head(9)
```

```python
# Total Stock Value

df['stockvalue'] = df['discountedsellingprice'] * df['availablequantity']
df[['stockvalue','discountedsellingprice','availablequantity']].head(9)
```

```python
# Category level Analysis

df.groupby('category').agg({
    'mrp': 'mean',
    'discountpercent': 'mean',
    'availablequantity': 'sum',
    'stockvalue': 'sum'
}).head(9)
```

```python
print("\n--- Average Selling Price by Category ---\n")
df.groupby('category')['discountedsellingprice'].mean()
```

```python
print("\n--- Total Quantity Sold by Category ---\n")
print(df.groupby('category')['quantity'].sum())
```

```python
# Highest Discount Product

df.sort_values(by='discountpercent', ascending=False).head(4)
```

```python
# Top expensive Product

df.sort_values('discountedsellingprice', ascending=False).head(4)
```

```python
print("\n--- Out of Stock Products ---\n")

print(df[df['outofstock'] == True][['name', 'category']].head(9))
```

```python
print("\n--- Low Available Quantity Products ---\n")
print(df[df['availablequantity'] <= 3][['name', 'availablequantity']].head(9))
```
